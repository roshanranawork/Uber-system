## 🗃️ Section 2 — Dataset Generation

We generate a realistic synthetic Uber dataset covering:
- **Riders** and **Drivers** with profiles
- **Rides** with timestamps, locations, fares, and statuses


```python
# ─── Helper Functions ───────────────────────────────────────────────────────

def generate_id(prefix, n=6):
    """Generate a unique ID like RID-A3F8B2"""
    return prefix + '-' + ''.join(random.choices(string.ascii_uppercase + string.digits, k=n))

def random_date(start_days_ago=365, end_days_ago=0):
    """Return a random datetime within the past N days"""
    delta = random.randint(end_days_ago, start_days_ago)
    return datetime.now() - timedelta(days=delta,
                                       hours=random.randint(0, 23),
                                       minutes=random.randint(0, 59))

random.seed(42)
np.random.seed(42)

# ─── Cities & Zones ─────────────────────────────────────────────────────────
CITIES = ['Ambala', 'Chandigarh', 'Delhi', 'Mohali', 'Panchkula', 'Kurukshetra']
ZONES  = ['Airport', 'City Centre', 'Railway Station', 'Mall', 'Residential', 'IT Hub', 'Hospital']
RIDE_TYPES = ['UberX', 'UberBlack', 'UberSUV', 'UberAuto']
RIDE_PURPOSES = ['Office Commute', 'Airport Drop', 'Shopping', 'Medical', 'Meeting', 'Leisure', 'Other']
STATUSES = ['Completed', 'Cancelled', 'In Progress']

print("✅ Configuration ready.")
print(f"   Cities  : {CITIES}")
print(f"   Zones   : {ZONES}")
print(f"   Ride Types: {RIDE_TYPES}")
```

```python
# ─── Generate Riders ────────────────────────────────────────────────────────
RIDER_NAMES = [
    'Roshan Rana', 'Priya Sharma', 'Amit Verma', 'Sneha Patel', 'Rahul Singh',
    'Kavita Joshi', 'Manish Gupta', 'Pooja Nair', 'Deepak Kumar', 'Anjali Mehta',
    'Vikram Yadav', 'Riya Kapoor', 'Sunil Tiwari', 'Neha Dubey', 'Arjun Reddy',
    'Simran Kaur', 'Tarun Bhatia', 'Divya Malhotra', 'Harsh Aggarwal', 'Meena Rawat'
]

riders = []
for i, name in enumerate(RIDER_NAMES):
    riders.append({
        'rider_id'    : generate_id('RDR'),
        'name'        : name,
        'email'       : name.lower().replace(' ', '.') + '@gmail.com',
        'phone'       : f"98{random.randint(10000000, 99999999)}",
        'city'        : random.choice(CITIES),
        'rating'      : round(random.uniform(3.5, 5.0), 1),
        'total_rides' : random.randint(5, 150),
        'member_since': (datetime.now() - timedelta(days=random.randint(30, 900))).strftime('%Y-%m-%d'),
        'wallet_balance': round(random.uniform(50, 2000), 2)
    })

riders_df = pd.DataFrame(riders)
print(f"✅ {len(riders_df)} Riders registered.")
display(riders_df.head(5))
```

```python
# ─── Generate Drivers ───────────────────────────────────────────────────────
DRIVER_NAMES = [
    'Ravi Kumar', 'Suresh Pal', 'Mohan Lal', 'Harish Sharma', 'Dinesh Negi',
    'Pawan Bisht', 'Vinod Thakur', 'Anil Chauhan', 'Rajesh Rawat', 'Sanjay Mehta',
    'Manoj Verma', 'Ramesh Gupta', 'Naresh Yadav', 'Trilok Singh', 'Bhupesh Tomar'
]
VEHICLES = ['Swift Dzire', 'Toyota Etios', 'Honda City', 'Hyundai Xcent',
            'Maruti Ertiga', 'Innova Crysta', 'Tata Nexon', 'Mahindra Scorpio']

drivers = []
for name in DRIVER_NAMES:
    drivers.append({
        'driver_id'     : generate_id('DRV'),
        'name'          : name,
        'phone'         : f"97{random.randint(10000000, 99999999)}",
        'vehicle'       : random.choice(VEHICLES),
        'ride_type'     : random.choice(RIDE_TYPES),
        'license_no'    : 'HR-' + str(random.randint(1000000, 9999999)),
        'city'          : random.choice(CITIES),
        'rating'        : round(random.uniform(3.8, 5.0), 1),
        'total_trips'   : random.randint(50, 800),
        'is_available'  : random.choice([True, True, True, False]),  # 75% available
        'joined'        : (datetime.now() - timedelta(days=random.randint(60, 1200))).strftime('%Y-%m-%d'),
        'earnings_total': round(random.uniform(5000, 80000), 2)
    })

drivers_df = pd.DataFrame(drivers)
print(f"✅ {len(drivers_df)} Drivers registered.")
display(drivers_df.head(5))
```

```python
# ─── Generate Rides ─────────────────────────────────────────────────────────
N_RIDES = 200

rides = []
for _ in range(N_RIDES):
    rider  = riders_df.sample(1).iloc[0]
    driver = drivers_df.sample(1).iloc[0]
    dist   = round(random.uniform(1.5, 45.0), 2)
    start  = random.choice(ZONES)
    end    = random.choice([z for z in ZONES if z != start])
    rtype  = driver['ride_type']
    dt     = random_date(365, 0)
    status = random.choices(STATUSES, weights=[70, 20, 10])[0]

    base_fare = {'UberX': 12, 'UberBlack': 20, 'UberSUV': 18, 'UberAuto': 8}
    per_km    = {'UberX': 9,  'UberBlack': 16, 'UberSUV': 14, 'UberAuto': 6}
    surge     = round(random.choice([1.0, 1.0, 1.2, 1.5, 2.0]), 1)
    fare      = round((base_fare[rtype] + per_km[rtype] * dist) * surge, 2)
    if status == 'Cancelled':
        fare = 0.0

    rides.append({
        'ride_id'      : generate_id('RID'),
        'rider_id'     : rider['rider_id'],
        'rider_name'   : rider['name'],
        'driver_id'    : driver['driver_id'],
        'driver_name'  : driver['name'],
        'ride_type'    : rtype,
        'pickup'       : start,
        'dropoff'      : end,
        'city'         : rider['city'],
        'distance_km'  : dist,
        'fare'         : fare,
        'surge_multiplier': surge,
        'status'       : status,
        'purpose'      : random.choice(RIDE_PURPOSES),
        'payment'      : random.choice(['Cash', 'UPI', 'Card', 'Wallet']),
        'rider_rating' : round(random.uniform(3.0, 5.0), 1) if status == 'Completed' else None,
        'driver_rating': round(random.uniform(3.5, 5.0), 1) if status == 'Completed' else None,
        'start_time'   : dt.strftime('%Y-%m-%d %H:%M:%S'),
        'end_time'     : (dt + timedelta(minutes=int(dist * 3))).strftime('%Y-%m-%d %H:%M:%S') if status != 'Cancelled' else None,
        'month'        : dt.strftime('%b'),
        'day'          : dt.strftime('%a'),
        'hour'         : dt.hour,
        'day_night'    : 'Day' if 6 <= dt.hour < 20 else 'Night'
    })

rides_df = pd.DataFrame(rides)
print(f"✅ {len(rides_df)} Rides generated.")
print(f"   Completed : {(rides_df.status=='Completed').sum()}")
print(f"   Cancelled : {(rides_df.status=='Cancelled').sum()}")
print(f"   In Progress: {(rides_df.status=='In Progress').sum()}")
display(rides_df.head(5))
```

## 💰 Section 4 — Fare Calculation Engine

Dynamic fare calculator with base fare, per-km rate, surge pricing, and ride-type tiers.


```python
# ─── Fare Engine ─────────────────────────────────────────────────────────────

FARE_TABLE = {
    #           base  per_km  min_fare
    'UberX'   : (12,   9,     40),
    'UberAuto': ( 8,   6,     25),
    'UberBlack': (20, 16,     80),
    'UberSUV' : (18,  14,     70),
}

def get_surge(hour):
    """Return surge multiplier based on time of day"""
    if   7 <= hour <= 9  : return 1.5   # Morning peak
    elif 17 <= hour <= 20: return 1.8   # Evening peak
    elif 23 <= hour or hour <= 5: return 1.3  # Late night
    else                 : return 1.0   # Normal

def calculate_fare(ride_type, distance_km, hour=None):
    """
    Calculate ride fare.
    Parameters:
      ride_type   : str  — UberX / UberAuto / UberBlack / UberSUV
      distance_km : float
      hour        : int (0-23), current hour for surge; defaults to now
    Returns:
      dict with fare breakdown
    """
    if ride_type not in FARE_TABLE:
        return {"error": f"Unknown ride type '{ride_type}'"}

    if hour is None:
        hour = datetime.now().hour

    base, per_km, min_fare = FARE_TABLE[ride_type]
    surge   = get_surge(hour)
    raw     = base + (per_km * distance_km)
    total   = max(round(raw * surge, 2), min_fare)
    tax     = round(total * 0.05, 2)
    grand   = round(total + tax, 2)

    return {
        'ride_type'     : ride_type,
        'distance_km'   : distance_km,
        'base_fare'     : base,
        'distance_fare' : round(per_km * distance_km, 2),
        'subtotal'      : round(raw, 2),
        'surge_mult'    : surge,
        'fare_after_surge': total,
        'tax_5pct'      : tax,
        'total_fare'    : grand,
        'hour'          : hour
    }

def print_fare(fare_dict):
    f = fare_dict
    print(f"┌─────────────────────────────────────┐")
    print(f"│         FARE ESTIMATE ({f['ride_type']})       │")
    print(f"├─────────────────────────────────────┤")
    print(f"│  Distance        : {f['distance_km']} km")
    print(f"│  Base Fare       : ₹{f['base_fare']}")
    print(f"│  Distance Fare   : ₹{f['distance_fare']}")
    print(f"│  Subtotal        : ₹{f['subtotal']}")
    print(f"│  Surge ({f['surge_mult']}x)      : ₹{round(f['fare_after_surge']-f['subtotal'],2)}")
    print(f"│  Tax (5%)        : ₹{f['tax_5pct']}")
    print(f"│  ─────────────────────────────────")
    print(f"│  TOTAL FARE      : ₹{f['total_fare']}")
    print(f"└─────────────────────────────────────┘")

# ── Demo ─────────────────────────────────────
print_fare(calculate_fare('UberX',    12.5, hour=8))   # morning surge
print()
print_fare(calculate_fare('UberBlack', 8.0, hour=14))  # normal hour
print()
print_fare(calculate_fare('UberAuto',  3.2, hour=18))  # evening surge
```

## 🚕 Section 5 — Book / Cancel a Ride

Core ride booking and cancellation engine with driver matching and wallet deduction.


```python
# ─── Booking System ──────────────────────────────────────────────────────────

def find_available_driver(city, ride_type):
    """Find nearest available driver in the same city for a given ride type"""
    pool = drivers_df[
        (drivers_df.city == city) &
        (drivers_df.ride_type == ride_type) &
        (drivers_df.is_available == True)
    ]
    if pool.empty:
        # fallback: any available driver
        pool = drivers_df[
            (drivers_df.ride_type == ride_type) &
            (drivers_df.is_available == True)
        ]
    if pool.empty:
        return None
    return pool.sample(1).iloc[0]

def book_ride(rider_id, ride_type, pickup, dropoff, distance_km, purpose='Other', payment='UPI'):
    """
    Book a ride for a rider.
    Returns ride_id if successful, None if failed.
    """
    global rides_df, drivers_df, riders_df

    # Validate rider
    rider_row = riders_df[riders_df.rider_id == rider_id]
    if rider_row.empty:
        print(f"❌ Rider ID '{rider_id}' not found.")
        return None
    rider = rider_row.iloc[0]

    # Find driver
    driver = find_available_driver(rider['city'], ride_type)
    if driver is None:
        print(f"❌ No available {ride_type} driver in {rider['city']} right now.")
        return None

    # Calculate fare
    hour = datetime.now().hour
    fare_info = calculate_fare(ride_type, distance_km, hour)
    fare = fare_info['total_fare']

    # Wallet check (for Wallet payment)
    if payment == 'Wallet' and rider['wallet_balance'] < fare:
        print(f"❌ Insufficient wallet balance. Required: ₹{fare}, Available: ₹{rider['wallet_balance']}")
        return None

    # Create ride record
    now = datetime.now()
    ride_id = generate_id('RID')
    new_ride = {
        'ride_id'         : ride_id,
        'rider_id'        : rider_id,
        'rider_name'      : rider['name'],
        'driver_id'       : driver['driver_id'],
        'driver_name'     : driver['name'],
        'ride_type'       : ride_type,
        'pickup'          : pickup,
        'dropoff'         : dropoff,
        'city'            : rider['city'],
        'distance_km'     : distance_km,
        'fare'            : fare,
        'surge_multiplier': fare_info['surge_mult'],
        'status'          : 'In Progress',
        'purpose'         : purpose,
        'payment'         : payment,
        'rider_rating'    : None,
        'driver_rating'   : None,
        'start_time'      : now.strftime('%Y-%m-%d %H:%M:%S'),
        'end_time'        : None,
        'month'           : now.strftime('%b'),
        'day'             : now.strftime('%a'),
        'hour'            : now.hour,
        'day_night'       : 'Day' if 6 <= now.hour < 20 else 'Night'
    }
    rides_df = pd.concat([rides_df, pd.DataFrame([new_ride])], ignore_index=True)

    # Mark driver as unavailable
    drivers_df.loc[drivers_df.driver_id == driver['driver_id'], 'is_available'] = False

    # Deduct wallet if needed
    if payment == 'Wallet':
        riders_df.loc[riders_df.rider_id == rider_id, 'wallet_balance'] -= fare

    print(f"✅ Ride Booked Successfully!")
    print(f"   Ride ID   : {ride_id}")
    print(f"   Rider     : {rider['name']}  →  Driver: {driver['name']}")
    print(f"   Route     : {pickup} ➜ {dropoff}  ({distance_km} km)")
    print(f"   Ride Type : {ride_type}  |  Fare: ₹{fare}  |  Surge: {fare_info['surge_mult']}x")
    print(f"   Payment   : {payment}  |  Status: 🟡 In Progress")
    return ride_id

def cancel_ride(ride_id, reason="User request"):
    """Cancel an In-Progress ride"""
    global rides_df, drivers_df

    ride_row = rides_df[rides_df.ride_id == ride_id]
    if ride_row.empty:
        print(f"❌ Ride ID '{ride_id}' not found.")
        return

    ride = ride_row.iloc[0]
    if ride['status'] != 'In Progress':
        print(f"⚠️  Ride '{ride_id}' is already '{ride['status']}' — cannot cancel.")
        return

    rides_df.loc[rides_df.ride_id == ride_id, 'status'] = 'Cancelled'
    rides_df.loc[rides_df.ride_id == ride_id, 'fare']   = 0.0

    # Re-mark driver as available
    drivers_df.loc[drivers_df.driver_id == ride['driver_id'], 'is_available'] = True

    print(f"🚫 Ride {ride_id} Cancelled.")
    print(f"   Reason : {reason}")
    print(f"   Driver {ride['driver_name']} is now available again.")

def complete_ride(ride_id, rider_rating=5.0, driver_rating=5.0):
    """Mark a ride as Completed"""
    global rides_df, drivers_df

    ride_row = rides_df[rides_df.ride_id == ride_id]
    if ride_row.empty:
        print(f"❌ Ride ID '{ride_id}' not found."); return

    ride = ride_row.iloc[0]
    if ride['status'] != 'In Progress':
        print(f"⚠️  Ride '{ride_id}' cannot be completed — status: {ride['status']}"); return

    end_time = datetime.now()
    rides_df.loc[rides_df.ride_id == ride_id, 'status']       = 'Completed'
    rides_df.loc[rides_df.ride_id == ride_id, 'end_time']     = end_time.strftime('%Y-%m-%d %H:%M:%S')
    rides_df.loc[rides_df.ride_id == ride_id, 'rider_rating'] = rider_rating
    rides_df.loc[rides_df.ride_id == ride_id, 'driver_rating']= driver_rating
    drivers_df.loc[drivers_df.driver_id == ride['driver_id'], 'is_available']  = True
    drivers_df.loc[drivers_df.driver_id == ride['driver_id'], 'total_trips']  += 1

    print(f"✅ Ride {ride_id} Completed!")
    print(f"   Fare Charged  : ₹{ride['fare']}")
    print(f"   Rider Rating  : ⭐ {rider_rating}  |  Driver Rating: ⭐ {driver_rating}")

# ── Demo ─────────────────────────────────────
sample_rider_id = riders_df.iloc[0]['rider_id']
booked_id = book_ride(sample_rider_id, 'UberX', 'City Centre', 'Airport', 14.5, 'Airport Drop', 'UPI')
```

```python
# ─── Ride History Functions ──────────────────────────────────────────────────

def rider_history(rider_id, last_n=10):
    """Show last N rides for a rider"""
    df = rides_df[rides_df.rider_id == rider_id].tail(last_n)
    if df.empty:
        print(f"No ride history for rider {rider_id}"); return
    cols = ['ride_id','ride_type','pickup','dropoff','distance_km','fare','status','start_time']
    print(f"📋 Ride History for {df.iloc[0]['rider_name']} (last {len(df)} rides):")
    display(df[cols].reset_index(drop=True))

def driver_history(driver_id, last_n=10):
    """Show last N rides for a driver"""
    df = rides_df[rides_df.driver_id == driver_id].tail(last_n)
    if df.empty:
        print(f"No ride history for driver {driver_id}"); return
    cols = ['ride_id','ride_type','pickup','dropoff','distance_km','fare','status','start_time']
    print(f"📋 Trip History for {df.iloc[0]['driver_name']} (last {len(df)} trips):")
    display(df[cols].reset_index(drop=True))

# ── Demo ─────────────────────────────────────
rider_history(riders_df.iloc[0]['rider_id'])
```

## 🔄 Section 7 — Real-Time Status Updates

Simulates ride lifecycle status transitions and driver location tracking.


```python
# ─── Status Engine ───────────────────────────────────────────────────────────
import time

RIDE_LIFECYCLE = [
    ("🔍 Searching for driver",    2),
    ("✅ Driver assigned",          1),
    ("🚗 Driver on the way",        3),
    ("📍 Driver arrived at pickup", 1),
    ("🚕 Ride in progress",         4),
    ("🏁 Approaching destination",  1),
    ("✅ Ride Completed",           0),
]

def simulate_ride_status(ride_id, fast=True):
    """
    Simulate real-time status updates for a ride.
    fast=True skips actual sleep for notebook demo.
    """
    ride_row = rides_df[rides_df.ride_id == ride_id]
    if ride_row.empty:
        print(f"❌ Ride {ride_id} not found."); return
    ride = ride_row.iloc[0]

    print(f"\n{'='*55}")
    print(f"  🚗 LIVE RIDE TRACKER — {ride_id}")
    print(f"  Rider  : {ride['rider_name']}")
    print(f"  Driver : {ride['driver_name']}")
    print(f"  Route  : {ride['pickup']} ➜ {ride['dropoff']}")
    print(f"  Fare   : ₹{ride['fare']}")
    print(f"{'='*55}")

    for status, wait in RIDE_LIFECYCLE:
        ts = datetime.now().strftime('%H:%M:%S')
        print(f"  [{ts}]  {status}")
        if not fast:
            time.sleep(wait)

    print(f"{'='*55}")
    print(f"  💳 Payment: {ride['payment']}  |  Amount: ₹{ride['fare']}")
    print(f"  Please rate your ride! ⭐⭐⭐⭐⭐")
    print(f"{'='*55}\n")

def get_ride_status(ride_id):
    """Quick status check for a ride"""
    row = rides_df[rides_df.ride_id == ride_id]
    if row.empty:
        print(f"❌ Ride {ride_id} not found."); return
    r = row.iloc[0]
    icon = {'Completed': '✅', 'Cancelled': '🚫', 'In Progress': '🟡'}.get(r['status'], '❓')
    print(f"{icon} Ride {ride_id}  |  Status: {r['status']}  |  Fare: ₹{r['fare']}  |  Driver: {r['driver_name']}")

# ── Demo ─────────────────────────────────────
# Simulate on a sample in-progress ride
ip_rides = rides_df[rides_df.status == 'In Progress']
if not ip_rides.empty:
    simulate_ride_status(ip_rides.iloc[0]['ride_id'], fast=True)
else:
    simulate_ride_status(rides_df.iloc[0]['ride_id'], fast=True)
```
